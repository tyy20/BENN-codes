```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

```

```python
x_train=pd.read_csv("./application-data-2/x_train.csv")
x_train=x_train.drop('Unnamed: 0', axis=1)
x_test=pd.read_csv("./application-data-2/x_test.csv")
x_test=x_test.drop('Unnamed: 0', axis=1)
y_train=pd.read_csv("./application-data-2/y_train.csv")
y_train=y_train.drop('Unnamed: 0', axis=1)
y_test=pd.read_csv("./application-data-2/y_test.csv")
y_test=y_test.drop('Unnamed: 0', axis=1)

y_suff_train=pd.read_csv("./application-results-2/y_suff_train_GSIR.csv")
y_suff_train=y_suff_train.drop('Unnamed: 0', axis=1)
y_suff_test=pd.read_csv("./application-results-2/y_suff_test_GSIR.csv")
y_suff_test=y_suff_test.drop('Unnamed: 0', axis=1)

```

```python
plt.plot(y_suff_train, y_train, 'o', color="C0")
plt.xlabel("Sufficient predicor")
plt.ylabel("Response")
plt.title("Results for training set")
plt.savefig("application-training-GSIR-2.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
plt.plot(y_suff_test, y_test, 'o', color="C1")
plt.xlabel("Sufficient predicor")
plt.ylabel("Response")
plt.title("Results for testing set")
plt.savefig("application-testing-GSIR-2.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
